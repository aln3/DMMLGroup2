# Clustering

## Imports & Data Preparation 

```python
import cv2
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import time

from numpy import genfromtxt
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import f_regression
from sklearn.metrics import accuracy_score
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.manifold import TSNE
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import mode
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering



num_pixels = 24
%matplotlib inline
```

```python
#Reading data from CSV files
x_train = genfromtxt('x_train_gr_smpl.csv', delimiter=',', skip_header =1)
y_train = genfromtxt('y_train_smpl.csv', delimiter=',', skip_header =1 )
```

```python
#Combine data with lable
y_train = y_train.reshape(12660,1)
data_combined = np.append(x_train, y_train, axis=1)
```

```python
#Randomises row order in reproducable manner
np.random.seed(0)
np.random.shuffle(data_combined)
```

```python
#Extract data and data lables that have undergone the same randomization
x_train = data_combined[:,:-1]
y_train = data_combined[:, [-1]].reshape(12660,)
```

## kMeans: No class attribute - Full Dataset


```python
# Adapted from https://jakevdp.github.io/PythonDataScienceHandbook/05.11-k-means.html

x = []
#Experiment to determine how the random seed affects the results produced by kMeans
t_start= time.time()
for i in range(10):
    
    #running kMeans algorithm with 10 centriods initilized at random with a maximum of 500 itterations.
    kmeans = KMeans(n_clusters=10, init='random', random_state=i, precompute_distances = True, max_iter = 500)
    #return clustering assignments for each instance.
    clusters = kmeans.fit_predict(x_train)
    
    #using boolean mask to match cluster assignment to lables 
    #create an array with the same structure as clusters but set all values to zero
    labels = np.zeros_like(clusters)
    # for each cluster lable
    for i in range(10):
        #if label matches set mask to true
        mask = (clusters == i)
        labels[mask] = mode(y_train[mask])[0]
    

    a = accuracy_score(y_train, labels)
    x.append(a)
    
t_finish = time.time()
#time taken to execute experiment in minutes
print((t_finish-t_start)/60)
```

```python
#Inspect x for highest accuarcy and make note of index, this  has a one-to-one correspondace with the random seed
print(x)
```

# Determine the Optimal K for K-Means

## The Elbow Method


```python

def calculate_WSS(points, kmax):
  sse = []
  for k in range(1, kmax+1):
    kmeans = KMeans(n_clusters = k).fit(points)
    centroids = kmeans.cluster_centers_
    pred_clusters = kmeans.predict(points)
    curr_sse = 0
    
    # calculate square of Euclidean distance of each point from its cluster center and add to current WSS
    for i in range(len(points)):
      curr_center = centroids[pred_clusters[i]]
      curr_sse += (points[i, 0] - curr_center[0]) ** 2 + (points[i, 1] - curr_center[1]) ** 2
      
    sse.append(curr_sse)
  return sse
```

```python
kmax = 40
possible_k = []

for i in range(1,kmax+1):
    possible_k.insert(i,i)


WWS = calculate_WSS(x_train_3,kmax)
#%matplotlib qt
%matplotlib inline
#Plotting wws vs k

plt.plot(possible_k, WWS)
```

## The Silhouette Method

```python
# Taken directly from [https://medium.com/analytics-vidhya/how-to-determine-the-optimal-k-for-k-means-708505d204eb]
sil = []
kmax = 40

# dissimilarity would not be defined for a single cluster, thus, minimum number of clusters should be 2
for k in range(2, kmax+1):
  kmeans = KMeans(n_clusters = k).fit(x_train_3)
  labels = kmeans.labels_
  sil.append(silhouette_score(x_train_3, labels, metric = 'euclidean'))
```

```python
# Plotting k vs score
possible_k = []
for i in range(2,kmax+1):
    possible_k.insert(i, i)

```

```python
%matplotlib qt
plt.plot(possible_k,sil)
```
